```python
#STEP: 4 ENCODING USING LABEL ENCODING AND GET DUMMIES



from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['ever_married'] = le.fit_transform(df['ever_married'])
df['Residence_type'] = le.fit_transform(df['Residence_type'])

df = pd.get_dummies(df, columns=['work_type', 'smoking_status'], drop_first=True)
df
```

```python
#STEP: 5 FILLING MISSING VALUES

from sklearn.impute import KNNImputer

# Apply KNN imputer (fill BMI missing values and any others)
imputer = KNNImputer(n_neighbors=5)
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Confirm missing values handled
print(df_imputed.isnull().sum())
```

```python
#STEP : 5 OUTLIER DETECTION
# List of numerical columns to check for outliers
numerical_cols = ['age','avg_glucose_level', 'bmi'] 

for col in numerical_cols:
    Q1 = df_imputed[col].quantile(0.25)
    Q3 = df_imputed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers
    outliers = df_imputed[(df_imputed[col] < lower_bound) | (df_imputed[col] > upper_bound)]
    outlier_count = outliers.shape[0]

    print(f"{col}: Found {outlier_count} outliers")

import seaborn as sns
import matplotlib.pyplot as plt

for col in numerical_cols:
    sns.boxplot(x=df_imputed[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Combine all numerical columns in one boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_imputed[numerical_cols], palette="pastel")
plt.title('Outlier Detection for Numerical Features', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
#Function to remove outliers using IQR
def remove_outliers(df, col):
    Q1 = df_imputed[col].quantile(0.25)
    Q3 = df_imputed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df_imputed[(df_imputed[col] >= lower_bound) & (df_imputed[col] <= upper_bound)]
    print(f"{col}: Removed {df_imputed.shape[0] - filtered_df.shape[0]} outliers")
    return filtered_df

# Apply outlier removal for each column
for col in numerical_cols:
    df_imputed = remove_outliers(df_imputed, col)

# Check final shape
print("Final dataset shape:", df_imputed.shape)

for col in numerical_cols:
    sns.boxplot(x=df_imputed[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
#STEP 7: SPLITTING DATA INTO TRAINING AND TESTING DATA
from sklearn.model_selection import train_test_split, GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
# Models without normalization needed

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB()
}

results = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"\nModel: {name}")
    print("Accuracy:", acc)
    print("Classification Report:\n", classification_report(y_test, y_pred))
    results[name] = acc

    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
   
    print(cm)

    from sklearn.metrics import ConfusionMatrixDisplay
    labels=[1,0]
    cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
    cmd.plot()

# # Plot accuracy comparison
# plt.figure(figsize=(10,6))
# sns.barplot(x=list(results.keys()), y=list(results.values()))
# plt.xticks(rotation=45)
# plt.title("Model Accuracy Comparison")
# plt.show()

```

```python
#STEP 8: NORMALIZING AND SCALING

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
# Models with normalization needed
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN": KNeighborsClassifier(),
    "SVC": SVC(),
    "MLP Classifier": MLPClassifier(max_iter=500)
}

results = {}

for name, model in models.items():
    model.fit(x_train_scaled, y_train)
    y_preds = model.predict(x_test_scaled)
    acc = accuracy_score(y_test, y_preds)
    print(f"\nModel: {name}")
    print("Accuracy:", acc)
    print("Classification Report:\n", classification_report(y_test, y_preds))
    results[name] = acc
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_preds)
   
    print(cm)

    from sklearn.metrics import ConfusionMatrixDisplay
    labels=[1,0]
    cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
    cmd.plot()

```
